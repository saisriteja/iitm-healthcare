```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from sympy import symbols
import sympy as sym
from IPython.display import Math,display
sym.init_printing()



class plotting():
    def __init__(self,k):
        self.p1 = np.random.randint(low=1, high=20, size=3)
        self.p2 = np.random.randint(low=1, high=20, size=3)
        self.d = self.p1-self.p2
        self.k = k
        self.x,self.y,self.z = symbols('X Y Z')
        self.x_points = np.zeros(self.k)
        self.y_points = np.zeros(self.k)
        self.z_points = np.zeros(self.k)
        self.distance = np.linalg.norm(self.p1-self.p2)
        self.cosa = 0
        self.cosb = 0
        self.cosc = 0
    def get_points(self):
        self.p = np.linspace(0,self.distance, num=self.k+2)*(1/self.distance)
        self.x_points = (self.d[0]*self.p[1:self.k+1])+(self.p2[0])
        self.y_points = (self.d[1]*self.p[1:self.k+1])+(self.p2[1])
        self.z_points = (self.d[2]*self.p[1:self.k+1])+(self.p2[2])
    def get_angles(self):
        self.cosa = np.degrees(np.arccos(self.d[0]/self.distance))
        self.cosb = np.degrees(np.arccos(self.d[1]/self.distance))
        self.cosc = np.degrees(np.arccos(self.d[2]/self.distance))
    def get_equation(self):
        display(Math(f'\\frac{{{sym.latex(self.x-self.p1[0])}}} {{{self.d[0]}}}'+'='+
                     f'\\frac{{{sym.latex(self.y-self.p1[1])}}} {{{self.d[1]}}}'+'='+
                     f'\\frac{{{sym.latex(self.z-self.p1[2])}}} {{{self.d[2]}}}'+'='+f'{{{sym.latex(self.k)}}}'))
                     
           
```

```python
s1 = plotting(10)
s1.get_points()
s1.get_angles()
x_5 = s1.x_points
y_5 = s1.y_points
z_5 = s1.z_points

points_1 = s1.p1
points_2 = s1.p2

x = [points_1[0],points_2[0]]
y = [points_1[1],points_2[1]]
z = [points_1[2],points_2[2]]


s1.get_equation()
```

```python

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x_5,y_5,z_5,'b',marker='^', s=40)

ax.scatter(x,y,z,'g',marker='8', s=30)
ax.plot(x, y, z, c='r')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
alpha,beta,gamma = symbols('alpha beta gamma')
ca = sym.cos(alpha)
cb = sym.cos(beta)
cg = sym.cos(gamma)
display(Math(f'{{{sym.latex(ca)}}} = {{{round(s1.cosa,2)}}}'))
display(Math(f'{{{sym.latex(cb)}}} = {{{round(s1.cosb,2)}}}'))
display(Math(f'{{{sym.latex(cg)}}} = {{{round(s1.cosc,2)}}}'))
```

```python
def fibanocci_number(n):
    golden_ratio = (1+(5**0.5))/(2)
    return (1/(5**0.5))*(golden_ratio**(n+1))

#https://math.stackexchange.com/questions/1145342/find-a-formula-for-the-nth-fibonacci-number


print("give the value of the nth number:")
n = int(input())
print(f"the {n}th fibanocci number is", round(fibanocci_number(n)))
```
